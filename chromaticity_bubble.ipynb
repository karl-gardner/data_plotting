# 1.1 Absorption Cycle in Chromaticity Plot

```python
# Plotting the *CIE 1931 Chromaticity Diagram*.
# The argument *standalone=False* is passed so that the plot doesn't get
# displayed and can be used as a basis for other plots.
fig, ax = colour.plotting.plot_chromaticity_diagram_CIE1931(standalone=False, show_spectral_locus=False)

for i, f in enumerate(os.listdir('/content/drive/MyDrive/nanofilm_data/data_files/chromaticity_pics/absorption')):
  RGB = plt.imread('/content/drive/MyDrive/nanofilm_data/data_files/chromaticity_pics/absorption/' + f)
  RGB = [np.mean(RGB[:,:,0]),np.mean(RGB[:,:,1]),np.mean(RGB[:,:,2])]
  XYZ = colour.sRGB_to_XYZ(RGB)
  xy =  colour.XYZ_to_xy(XYZ)
  # Plotting the *CIE xy* chromaticity coordinates.
  x, y = xy
  ax.plot(x, y, 'o-', color='black', markersize=4)
  # Annotating the plot.
  # if i == 11:
  #   ax.plot(x, y, 'o-', color='blue', markersize=4)
  #   ax.annotate(f[:-4],xy=xy,xytext=(-70, 60),textcoords='offset points',arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=-0.2'))

ax.set_title('Absorption Cycle', fontsize=25)
ax.set_xlabel('CIE X', fontsize=20)
ax.set_ylabel('CIE Y', fontsize=20)
ax.tick_params(axis='both', which='major', length=14,width=3.4,labelsize=15)
ax.set_xlim(-0.15,0.85)
ax.set_ylim(-0.07,0.9)
plt.close(fig)
fig.savefig('/absorption.png',dpi=600, bbox_inches='tight')
# files.download('/chromaticity.png')


# Add a border with cv2
funcs.add_border(im_path = "/absorption.png")
files.download('/absorption.png')

Image('/absorption.png', width=600, height=600)
```
